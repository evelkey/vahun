```python
corp_path='/home/velkey/corp/webkorpusz.wpl'
import tensorflow as tf
import keras
import scipy
import numpy as np
import time
```

```python
def is_hun_word(word):
    simbol='0123456789-,;.!?:’\”\"/\\|_@#$%^&*~`+ =<>()[]{}'
    accents = 'áéíóöőúüű'
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    hun_word=True
    if "eeeoddd" in word:
        return False
    for char in simbol:
        if char in word:
            return False
    return True
```

```python
def read_line_wpl(line,only_words=False,lower=False):
    if ( not only_words):
        if (not lower):
            return line.replace("\n","")
        else:
            return line.lower().replace("\n","")
    else:
        if(is_hun_word(line.lower())):
            if(not lower):
                return line.replace("\n","")
            else:
                return line.lower().replace("\n","")
        else:
            return ""
```

```python
def read_words(size,uniqe=True,only_words=True,lowercase=True):
    i=0
    if uniqe:
        batch=set()
    else: 
        batch=[]
    start=time.time()
    with open(corp_path) as f:
        for line in f:
            if len(batch)==size:
                break
            else:
                i+=1
                if i%1000000==0: 
                    if i!=0: print("Reading file, speed: ",1000000/(time.time()-start)," words/s")
                    start=time.time()
                if uniqe:
                    batch.add(read_line_wpl(line,only_words,lowercase))
                else:
                    batch.append(read_line_wpl(line,only_words,lowercase))
    return list(batch)
```

```python
batch=read_words(10000,True)
print(batch[0:40])
```

```python
def featurize_data(x, maxlen):

    simbol='0123456789-,;.!?:’\”\"/\\|_@#$%^&*~`+ =<>()[]{}'
    accents = 'áéíóöőúüű'
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    
    hun_alphabet=alphabet+accents
    
    feature_tensor = np.zeros((len(x), maxlen, len(hun_alphabet)))
    for dix,item in enumerate(x):
        counter = 0
        one_hot = np.zeros((maxlen, len(hun_alphabet)))
        chars = list(item.lower())
        for i in range(len(chars)):
            if len(chars)>maxlen:
                break
            else:
                if chars[i] in hun_alphabet:
                    one_hot[maxlen-len(chars)+i,hun_alphabet.find(chars[i])]=1
                
            
           
        feature_tensor[dix, :, :] = one_hot

    return feature_tensor

def get_word(x,maxlen):
    simbol='0123456789-,;.!?:’\”\"/\\|_@#$%^&*~`+ =<>()[]{}'
    accents = 'áéíóöőúüű'
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    hun_alphabet=alphabet+accents
    out=""
    for i in range(maxlen):
        if x[i,:].any()==1:
            out+=hun_alphabet[x[i,:].argmax()]
    return out
```

```python
train=featurize_data(batch[0:int(len(batch)*0.8)],30)
test=featurize_data(batch[int(len(batch)*0.8):len(batch)],30)
```

```python
a=featurize_data(["","eeeefsdf"],30)
print(a[1].sum())
get_word(a[1],30)
```

```python
x_train = train.reshape((len(train), np.prod(train.shape[1:])))
x_test = test.reshape((len(test), np.prod(test.shape[1:])))
print(x_train.shape)
```

```python
from keras.layers import Input, Dense
from keras.models import Model

from keras import regularizers

encoding_dim = 100

input_img = Input(shape=(30*35,))
# add a Dense layer with a L1 activity regularizer
encoded = Dense(encoding_dim, activation='relu')(input_img)
decoded = Dense(30*35, activation='sigmoid')(encoded)

autoencoder = Model(input=input_img, output=decoded)

# this model maps an input to its encoded representation
encoder = Model(input=input_img, output=encoded)

# create a placeholder for an encoded (32-dimensional) input
encoded_input = Input(shape=(encoding_dim,))
# retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]
# create the decoder model
decoder = Model(input=encoded_input, output=decoder_layer(encoded_input))

autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python
autoencoder.fit(x_train, x_train,
                nb_epoch=1000,
                batch_size=256,
                shuffle=False,
                validation_data=(x_test, x_test))
```

```python
# use Matplotlib (don't ask)
import matplotlib.pyplot as plt
encoded_imgs=encoder.predict(x_train)
decoded_imgs = decoder.predict(encoded_imgs)
n = 10  # how many digits we will display
plt.figure(figsize=(20, 4))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(30, 35))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(30,35))

    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
